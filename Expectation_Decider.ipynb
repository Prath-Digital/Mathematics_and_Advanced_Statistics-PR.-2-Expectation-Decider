This notebook analyzes **probability patterns** in student performance data 📚
to predict the likelihood of passing a **competitive mathematics exam** 🧠✨.

## 🎯 Project Objective

Analyze a dataset of **200 students** using probability concepts such as:
- 📈 Empirical probability
- 🎲 Random variables
- 📊 Probability distributions
- 🔗 Conditional probability
- 🧮 Bayes Theorem

```python
import pandas as pd

data = pd.read_csv("./data/student_performance_data.csv")

n = len(data)

data.head()
```

## 🧠 1. Understanding Probability

Probability measures the **chance of an event occurring** 🎲 and ranges
from **0 (impossible)** ❌ to **1 (certain)** ✅.

```python
passed = (data["final_exam_pass"] == "Pass").sum()
empirical_probability = passed / n
empirical_probability
```

### 📌 Empirical Probability Result

This value represents the **real-world probability** 📊 of a student
passing the exam based on observed data.

## 🎲 2. Random Variable & Probability Distribution

Let **X** be the number of students who pass when **3 students** are selected randomly 👥.

This follows a **Binomial Distribution** 📈.

```python
from math import comb

p = empirical_probability
n_trials = 3

probability_distribution = {
    k: comb(n_trials, k) * (p**k) * ((1-p)**(n_trials-k))
    for k in range(4)
}

probability_distribution
```

```python
mean = n_trials * p
variance = n_trials * p * (1 - p)

mean, variance
```

### 📐 Mean & Variance

These values describe the **expected outcome** 🎯 and **spread** of results 📊.

## 🔵 3. Venn Diagram Conditions

- ⏱️ Study hours **> 10 per week**
- 🏫 Attendance **> 80%**

The overlap represents students satisfying **both conditions** ✅.

```python
import matplotlib_venn._venn2 as venn

study_10 = (data["study_hours"] > 10).sum()
attendance_80 = (data["attendance"] > 80).sum()
both_conditions = ((data["study_hours"] > 10) & (data["attendance"] > 80)).sum()

study_10_not_attendance_80 = study_10 - both_conditions
attendance_80_not_study_10 = attendance_80 - both_conditions

venn.venn2(subsets=(study_10_not_attendance_80, attendance_80_not_study_10, both_conditions), set_labels=('Study > 10 hrs', 'Attendance > 80%'))
```

## 📋 4. Contingency Table Analysis

```python
contingency_table = pd.crosstab(
    data["group_discussion"],
    data["final_exam_pass"],
    margins=True
)

contingency_table
```

```python
joint_probability = contingency_table.loc["Yes", "Pass"] / n
marginal_probability = contingency_table.loc["All", "Pass"] / n
conditional_probability = contingency_table.loc["Yes", "Pass"] / contingency_table.loc["Yes", "All"]

joint_probability, marginal_probability, conditional_probability
```

## 🧮 5. Bayes Theorem Application

Using historical data 📚 to calculate:

**P(Pass | High Attendance)** 🎓

```python
bayes_probability = (joint_probability * marginal_probability) / conditional_probability
bayes_probability
```

## ✅ Final Conclusion 🎉

Students are **more likely to pass** when they:
- 📖 Study consistently
- 🏫 Maintain high attendance
- 🗣️ Participate in group discussions

Probability and Bayes analysis confirm these factors are **strong predictors** 🚀.
